```python

for i in worktable:
   
    
    yr = worktable[i]['created']
    if len(yr) == 6:
        yr = yr.split('.')
        if len(yr) == 2:
            worktable[i]['created'] = str(yr[-1])
            continue
        else:
            continue
    
    if len(yr) > 9:
        if len(yr) == 10:
            yr = yr.split('.')
  #          print(yr)
            if len(yr) > 2:
                worktable[i]['created'] = str(yr[2])
                continue
           
            else:
                continue
            

        else:    
            continue

            
    if len(re.findall('-',yr))>0:
        
        yr = yr.split('-')

#        print(i)
#        print(yr)
        for k in range(len(yr)):
            yr[k] = int(yr[k])
   
        worktable[i]['created'] = str(statistics.mean(yr))

    
```

```python
periods = [1950,1960,1970,1980,1990,2000,2010]
per_dict = {}
per_count = {}
for j in periods:
    per_dict[j] = []
    
for x in periods:
    per_count[x] = {'w_count_tbs' : 0}
for item in worktable:
    try:
        period = min(periods, key=lambda x:abs(x-int(worktable[item]['created'])))
        if per_count[period]['w_count_tbs'] < 15000000:
            if worktable[item]['words'] == 'none':
                continue
            per_count[period]['w_count_tbs'] += int(worktable[item]['words'])
            per_dict[period].append(item)
    except:
  #      print(worktable[item]['created'] + ' ??')
        continue
    
    
    
    
    
```
